### HR Analytics Employee Attrition Dataset

### Introduction

It is true that every year most of the companies hire various employees for their company , through this there is a huge cost involved in hiring employees and training those employees. We must have heard that when some of the company hire an employee some of those will have to sign the bond before joining, it is to maintain and secure the cost involve in training the new employees. This is where HR Analytics comes in to the picture in order to improve the performance of the employee , better return on investment, also provide an opportunity to mandate and imporve the current process.

Attrition in HR and how its Affecting in the Companies¶
Attrition here is the loss to the companies, It is verry important to maintain the work life balance and motivation level of the employees in order to retain the good and relevant employees of the company.

Attrition is a process which involves in increasing the organizational cost for any company , starting with job posting ,hiring process, paperwork, hire new trainer(trainer cost) etc..

Thus, I am here going to analyse this special dataset of HR analytic which consist of 1470 rows and 35 columns with different attributes

### Splitting the dataset into independent and target variable

```python
from sklearn.model_selection import train_test_split
x = df1.drop(['Attrition'], axis = 1)
y=df1["Attrition"]
```

### Handling Class Imbalanced problem by oversampling the minority class

```python
from imblearn.over_sampling import SMOTE
SM =SMOTE()
x_over, y_over = SM.fit_resample(x, y)
```

```python
y_over.value_counts()
```

```python
sn.countplot(y_over)
```

Now the each class is having same number of '0' and '1'

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, KFold
from sklearn.metrics import roc_auc_score,roc_curve,accuracy_score,precision_score,confusion_matrix,recall_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.linear_model import LogisticRegression
MaxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x_over,y_over,test_size=0.3,random_state=i)
    mod=LogisticRegression()
    mod.fit(x_train,y_train)
    pred=mod.predict(x_test)
    acc=accuracy_score(y_test,pred)
    if acc>MaxAccu:
        MaxAccu=acc
        maxRS=i
print("Best accuracy is",MaxAccu,'on random state',maxRS)
```

### Creating train_test split 

```python
# Splitting the dataset into trianing and testing data
x_train,x_test,y_train,y_test=train_test_split(x_over,y_over,test_size=0.3,random_state=maxRS)
```

### Perform Standard Scaler Tecnique

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
#Using logistic regression 
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression()
classifier.fit(x_train,y_train)
pred=classifier.predict(x_test)
lr_acc=accuracy_score(y_test,pred)
print('Accuracy score using logistic regression model is',lr_acc*100)
lr_cv=cross_val_score(classifier,x,y,cv=5).mean()
print("Cross validation score",lr_cv*100)
print('-----------------------------')
lr_acc1=classification_report(y_test,pred)
print(lr_acc1)
con_mat = confusion_matrix(y_test,pred)
print('-----------------------------')
print('\n Confusion Matrix')
plt.figure(figsize= (6,4))
sn.heatmap(con_mat, annot = True,fmt='d',cmap="coolwarm")
```

```python
#Using random forest classification
from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier() # creating instance of class
RF.fit(x_train,y_train) #Fitting the data
pred=RF.predict(x_test) # Predicting the data 
rfc_acc=accuracy_score(y_test,pred)
print("Accuracy of Random Forest Classificarion is ",rfc_acc*100)
rfc_cv=cross_val_score(RF,x,y,cv=5).mean()
print("Cross validation score",rfc_cv*100)
print('-------------------------------')
acc=classification_report(y_test,pred)
print(acc)
con_mat = confusion_matrix(y_test,pred)
print('-------------------------------')
print('\n Confusion Matrix')
plt.figure(figsize= (6,4))
sn.heatmap(con_mat, annot = True,fmt='d',cmap="YlGnBu")
```

```python
#Using Support vector classification
from sklearn.svm import SVC
svc=SVC(kernel='rbf',random_state=132)
svc.fit(x_train,y_train)
pred=svc.predict(x_test)
svc_acc=accuracy_score(y_test,pred)
print("Accuracy of Support Vector Classification is ",svc_acc*100)
svc_cv=cross_val_score(svc,x,y,cv=5).mean()
print("Cross validation score",svc_cv*100)
print('-------------------------------')
acc=classification_report(y_test,pred)
print(acc)
con_mat = confusion_matrix(y_test,pred)
print('-------------------------------')
print('\n Confusion Matrix')
plt.figure(figsize= (6,4))
sn.heatmap(con_mat, annot = True,fmt='d',cmap="YlGnBu")
```

```python
### using gradient boosting classifier
from sklearn.ensemble import GradientBoostingClassifier
clf=GradientBoostingClassifier(n_estimators=300,random_state=0)
clf.fit(x_train, y_train)
pred=clf.predict(x_test)
gbc_acc=accuracy_score(y_test,pred)
print('Accuracy score using gradient boosting classifier model is',gbc_acc*100)
gbc_cv=cross_val_score(clf,x,y,cv=5).mean()
print("Cross validation score",gbc_cv*100)
print('-------------------------------')
gbc_acc1=classification_report(y_test,pred)
print(gbc_acc1)
con_mat = confusion_matrix(y_test,pred)
print('-------------------------------')
print('\n Confusion Matrix')
plt.figure(figsize= (6,4))
sn.heatmap(con_mat, annot = True,fmt='d',cmap="coolwarm")
```

```python
model=['Logistic Regression','Random Forest Classifier','Support Vector Classifier','Gradient boosting Classifier']

acc=[lr_acc*100,rfc_acc*100,svc_acc*100,gbc_acc*100]

cv_score=[lr_cv*100,rfc_cv*100,svc_cv*100,gbc_cv*100]
diff=[acc[0]-cv_score[0],acc[1]-cv_score[1],acc[2]-cv_score[2],cv_score[3]-acc[3]]
diff
result=pd.DataFrame({'Model':model,'Accuracy_score':acc,'Cross Validation_score':cv_score,'Difference':diff})
result
```
